```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("data/lotto_history.csv", skiprows=1, usecols=(1,2,3,4,5,6,7), delimiter=',', dtype=int)

```

```python
# count the number of times each number is drawn
```

```python

# Create a histogram
plt.hist(data, bins=20, color='blue', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution Chart')

# Show the chart
plt.show()
```
